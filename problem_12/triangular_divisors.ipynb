## Problem 12

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
@njit
def trian_divisors(n_div = 500):
    
    # == constructing triangle series == #
    
    k = 1.0
    j = 1.0
    
    n = 0
    
    while n <= n_div:

        j = j + 1.0
        k = k + j
        
        # == finding divisors == #
        
        divisors = np.zeros(int(2 * np.sqrt(k)))

        for i in prange(1, np.sqrt(k)+1):
            if k % i == 0:
                if k/i == i:
                    divisors[i-1] = int(i)
                else:
                    divisors[i-1] = int(i)
                    divisors[-i] = int(k/i)

        divisors = divisors[np.nonzero(divisors)]
        
        n = divisors.size
        
    
    return k

trian_divisors(5)
```

```python
%time
trian_divisors(500)
```
